```python
# Import the pandas module
import pandas as pd
# Import the DataFrame
df = pd.read_csv('/content/Test180.csv', skiprows=19)
test_dr1 = df.drop(['Cell','Test','Rack','Shelf','Position','Cell ID', 'Step'], axis=1).fillna(method="bfill")
test_drop1 = test_dr1.fillna(method='ffill')
test_drop1
# Print the DataFrame
test_drop1
```

```python
from matplotlib import pyplot as plt

# Assuming test_drop1 is your DataFrame
test_drop1['Charge Capacity (mAh)'].plot(kind='line', figsize=(8, 4), title='Charge Capacity (mAh)')

# Add grid to the plot
plt.gca().grid(True, which='both', linestyle='--', linewidth=0.5)


plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns


def _plot_series(series, series_name, series_index=0):
    palette = list(sns.palettes.mpl_palette('Dark2'))
    xs = series['Total Time (Seconds)']
    ys = series['TC1 ']

    plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])


fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = test_drop1.sort_values('Total Time (Seconds)', ascending=True)
_plot_series(df_sorted, '')

# Add grid to the plot
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.xlabel('Total Time (Seconds)')
_ = plt.ylabel('TC1 ')
```
